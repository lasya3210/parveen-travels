```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python

customers = pd.read_csv("customers.csv") 
```

```python
routes = pd.read_csv("routes.csv") 
```

```python
bookings=pd.read_csv("bookings.csv")
```

```python
#BOOKINGS_DATA
# Plot: Number of bookings per Route 
plt.figure(figsize=(17, 5))
sns.countplot(data=bookings, x="Route_ID", order=bookings["Route_ID"].value_counts().index,color='skyblue')
plt.title("Number of Bookings per Route")
plt.xlabel("Route ID")
plt.ylabel("Number of Bookings")
plt.xticks(rotation=45)
plt.show()
```

```python
#CUSTOMERS_DATA
plt.figure(figsize=(8, 5))
sns.countplot(data=customers, x="Gender", palette="coolwarm",hue="Gender")
plt.title("Gender Distribution of Customers")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python


plt.figure(figsize=(6, 5))
sns.boxplot(y=bookings["Rating"], color="lightblue")  # Use 'y=' for a vertical boxplot
plt.title("Distribution of Ratings")
plt.ylabel("Ratings")
plt.show()
```

```python
#CUSTOMERS_DATA
# Plot: Count Plot of Age Group Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=customers, x="Age_Group", palette="viridis",hue="Age_Group")
plt.title("Age Group Distribution of Customers")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()
```

```python
#CUSTOMERS_DATA
# Plot : Preferred Bus Type Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=customers, y="Preferred_Bus_Type", palette="Set2",hue="Preferred_Bus_Type")
plt.title("Preferred Bus Types Among Customers")
plt.xlabel("Count")
plt.ylabel("Bus Type")
plt.show()
```

```python
#ROUTES_DATA
# Plot: Distance vs Ticket Price
plt.figure(figsize=(8, 5))
sns.scatterplot(data=routes, x="Distance (km)", y="Ticket_Price (INR)", hue="Bus_Type", palette="Dark2")
plt.title("Distance vs Ticket Price by Bus Type")
plt.xlabel("Distance (km)")
plt.ylabel("Ticket Price (INR)")
plt.show()
```

```python
#ROUTES_DATA
# Bus Type Distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='Bus_Type', data=routes, palette='coolwarm', hue='Bus_Type')
plt.title('Bus Type Distribution')
plt.xlabel('Bus Type')
plt.ylabel('Count of Routes')
plt.xticks(rotation=45)
plt.show()
```

```python
#ROUTES_DATA
#  Ticket Price vs. Duration (Scatter Plot using Matplotlib)
plt.figure(figsize=(8, 5))
plt.scatter(routes['Duration (hours)'], routes['Ticket_Price (INR)'])
plt.title('Ticket Price vs. Duration')
plt.xlabel('Duration (hours)')
plt.ylabel('Ticket Price (INR)')
plt.show()
plt.show()
```

```python
#ROUTES_DATA
#  Bus Type vs. Duration (Box Plot using Seaborn)
plt.figure(figsize=(8, 5))
sns.boxplot(x='Bus_Type', y='Duration (hours)', data=routes, palette='rainbow',hue='Bus_Type')
plt.title('Bus Type vs. Duration')
plt.xlabel('Bus Type')
plt.ylabel('Duration (hours)')
plt.show()
```
